```python
import pyarrow.parquet as pq
df=pq.read_table(source='C://Users//kk//Downloads//yellow_tripdata_2022-01.parquet').to_pandas()
```

```python
df.reset_index(inplace=True)
df.set_index('VendorID',inplace=True)
df.dropna(axis=0,how='any',inplace=True)
df.drop('index',axis=1,inplace=True)
df.rename(columns={'passenger_count':'passanger_count','airport_fee':'airport_free'},inplace=True)
df.sort_values(by=["tpep_pickup_datetime"],inplace=True)
df.head(5)
```
